```python
import matplotlib.pyplot as plt

# Loss values
loss = [1.021, 0.965, 0.952, 0.922, 0.915, 0.858, 0.780, 0.764, 0.768, 0.728, 0.721, 0.681, 0.737, 0.716]

# Epochs
epochs = range(1, len(loss) + 1)

# Creating the plot
plt.figure(figsize=(10, 6))  # Optional: Adjusts the figure size
plt.plot(epochs, loss, marker='o', linestyle='-', color='blue', label='Training Loss')
plt.title('TGN Training Loss per Epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.grid(True)  # Optional: Adds a grid for better readability
plt.legend()
plt.show()
```
